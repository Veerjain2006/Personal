# **Section 5: Data Science**

**I am choosing to stay with my question that I have in Section 4: if there was a huge difference between the time before the elections and after on the different sections of the economy. For this I will use a linear regression prediction model and find the difference between the predicted value and the actual value and proceed to make a scatter plot on the residual value.**

```python
#creating a predicition model
# changing the data to accomadate exact dates as the independant variable
import pandas as pd
from sklearn.linear_model import LinearRegression
df_new = pd.read_csv("DS Project #2 - Sheet1 (4).csv")
df_new.drop("Unnamed: 0", axis=1, inplace= True)
model = LinearRegression()
ind = ["Date_numeric"]
dep = "TSLA"
model = model.fit(df_new[ind],df_new[dep])
df_new["Predicted_values_TSLA"] = model.predict(df_new[ind])
ind = ["Date_numeric"]
dep = "BTC"
model = model.fit(df_new[ind],df_new[dep])
df_new["Predicted_values_BTC"] = model.predict(df_new[ind])
ind = ["Date_numeric"]
dep = "DOW JONES"
model = model.fit(df_new[ind],df_new[dep])
df_new["Predicted_values_DOW"] = model.predict(df_new[ind])
ind = ["Date_numeric"]
dep = "GOLD"
model = model.fit(df_new[ind],df_new[dep])
df_new["Predicted_values_GOLD"] = model.predict(df_new[ind])

```

```python
#Finding the residual of the actual price vs the predicted price

df_new["residual_TSLA"] = df_new["TSLA"] - df_new["Predicted_values_TSLA"]
df_new["residual_BTC"] = df_new["BTC"] - df_new["Predicted_values_BTC"]
df_new["residual_DOW"] = df_new["DOW JONES"] - df_new["Predicted_values_DOW"]
df_new["residual_GOLD"] = df_new["GOLD"] - df_new["Predicted_values_GOLD"]
df_new
```
